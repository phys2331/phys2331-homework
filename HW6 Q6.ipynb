# Part 1 Solution

The question is, What is the sign of the mobile charges, and which way do they move?

To answer this question, what we really need is to determine what the charge distribution looks like on the bar.  Unfortunately, they don't directly give us this information.  Instead, they tell us the sign of the voltage.  So first we need to use that information to determine where the charge is on the bar, and then we can figure out what the sign of the mobile charges is and what direction they're moving.

## Charge distribution on the bar

$\Delta V2$ is positive, and the book gives us a hint: <mark>Remember that a voltmeter gives a positive reading if the positive lead is connected to a higher potential location than the negative lead.</mark>.  I prefer to think of a voltmeter as telling us the direction of the electric field.  The negative lead (black) is the start of the path, and the positive lead (red) is the end of the path.  Because $\Delta V = -\int\vec{E}\cdot\vec{dl}$, if the path goes against the electric field, the voltage will be positive and if the path goes with the electric field, the voltage will be negative.  So in the case of $\Delta V2$, that means the electric field must be pointing up, which means there must be positive charge along the bottom of the bar and negative charge on the top,

$\Delta V1$ is also telling us the direction of the electric field, this time along the bar.  Because $\Delta V1$ is negative, the electric field must point along the path - to the right.



## Now we know the direction of the electric fields

So now we know the electric has a component to the right that drives the charges across the bar and component upwards that's formed as the charge accumulated on the top and bottom of the bar.  Because $\vec{F} = q\vec{E}$, the component to the right tells us that we either have positive charges moving to the right or negative charges moving to the left.

Let's consider each of these, and see which one will result in positive charges accumulating on the bottom of the bar.

## Is it negative charges moving to the left?

For negative charges moving to the left the picture looks like:

![hall_negative.png](images/hall_negative.png)

We can use $\vec{F} = q\vec{v}\times\vec{B}$ to determine the force on the negative charges due to the magnetic field, and see if they end up being pushed upwards.  We have that $q$ is $-e$, the charge of the electron.  Looking at this image, we see that $\vec{v} = \langle -v, 0, 0\rangle$.  The magnetic field $B$ is represented with a dot, which means it's coming out of the page, or $\vec{B} = \langle 0, 0, B\rangle$.  Now we can use these values to calculate the direction of the force.

```python
# sympy is a python library that allows symbolic manipulation
# similar to Mathematica or symbolab
# In my opinion, Mathematica documentation is outstanding,
# so if you want symbolic manipulation and have Mathematica installed,
# use Mathematica!
# Here I'll use sympy so we can see the result in our notebook.
import sympy as s

e, v, B = s.symbols("e, v, B")
v_vec = s.Matrix([-v, 0, 0])
B_vec = s.Matrix([0, 0, B])

Force_vec = -e*v_vec.cross(B_vec)

# The result is a vector that points downward
print (Force_vec)
```

Since the force on the negative charges is downward, these negative charges would get pushed to the bottom of the bar.  But that's not what we see from the voltage - we instead see positive charges on the bottom.  So this isn't the right answer.

We only have two options - negative charges moving to the left or positive charges moving to the right.  So the answer must be positive charges moving to the right.  Let's look at that diagram and confirm that the positive charges would in fact be pushed to the bottom as expected.

## Is it positive charges moving to the right?

For positive charges moving to the right the picture looks like:

![hall_right.png](images/hall_right.png)

We can use $\vec{F} = q\vec{v}\times\vec{B}$ to determine the force on the negative charges due to the magnetic field, and see if they end up being pushed upwards.  We have that $q$ is $+e$, opposite the charge of the electron (solid state physicists call these "holes").  Looking at this image, we see that $\vec{v} = \langle v, 0, 0\rangle$.  The magnetic field $B$ is represented with a dot, which means it's coming out of the page, or $\vec{B} = \langle 0, 0, B\rangle$.  Now we can use these values to calculate the direction of the force.

```python
import sympy as s

e, v, B = s.symbols("e, v, B")
v_vec = s.Matrix([v, 0, 0])
B_vec = s.Matrix([0, 0, B])

Force_vec = e*v_vec.cross(B_vec)

# The result is a vector that points downward
print (Force_vec)
```

The positive charges would feel a downward force, and get pushed to the bottom!  That's what we think the charge distribution looks like, so that's the answer!  Hooray!
