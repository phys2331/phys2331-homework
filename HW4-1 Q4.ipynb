# 4. Reduce $\vec{E}\cdot\vec{dl}$ to a scalar quantity by correctly taking the dot product

There are two good ways to do this!  

## Use $\vec{a}\cdot\vec{b} = |\vec{a}|*|\vec{b}|*\cos\theta$ - no coordinate system needed

The symbol $\theta$ here means the angle between the two vectors.  Since I am taking $\vec{E}\cdot\vec{dl}$, the angle between the Electric field and the path is 180 degrees in between the plates.   So my dot product looks like

$\vec{E}\cdot\vec{dl} = |\vec{E}|*|\vec{dl}|*\cos 180 = -E*dl$


Where in the final step I've dropped the vector symbols to indicate that these are magnitudes.  We often do this in physics even though it's less clear!

## Use $\vec{a}\cdot\vec{b} = a_x*b_x + a_y*b_y+ a_z*b_z$ - coordinate system needed!

With this method, we will use the coordinate system described in the problem: <mark>Assume the  +ùë• axis is to the right, the  +ùë¶ axis is up, and the  +ùëß  axis is out</mark>

And then we can write our vectors using that coordinate system.  Between the plates, we have

$\vec{E}\cdot\vec{dl} = \langle \frac{Q/A}{\epsilon_0}, 0, 0\rangle \cdot \langle dx, dy, dz\rangle = \frac{Q/A}{\epsilon_0} dx + 0 dy + 0 dz =\frac{Q/A}{\epsilon_0} dx $

# 5. Correctly evaluate the integral

The voltage is $\Delta V = -\int\vec{E}\cdot\vec{dl}$.  We've now got some good expressions for $\vec{E}\cdot\vec{dl}$, so let's integrate them!

## Use $\vec{a}\cdot\vec{b} = |\vec{a}|*|\vec{b}|*\cos\theta$ - no coordinate system needed

In this case we had $\vec{E}\cdot\vec{dl} = -E*dl$, where $E$ and $dl$ are magnitudes of the vectors $\vec{E}$ and $\vec{dl}$.  So

$\Delta V = -\int-E*dl$

The electric field in between two large plates is very nearly constant (especially near the middle).  So the magnitude of the electric field is also very nearly constant.  So I can pull $E$ out of the integral, and also multiply -1*-1:

$\Delta V = E\int dl$

The integral of $dl$ is the sum of all the tiny steps I make when walking from A to B - in other words, the distance (a magnitude) from A to B.  So

$\Delta V = E*(distance\,from\,A\,to\,B)$

In this case we had In this case we had $\vec{E}\cdot\vec{dl} = \frac{Q/A}{\epsilon_0} dx$.  So

$\Delta V = -\int \frac{Q/A}{\epsilon_0} dx$

We can pull out the constants:

$\Delta V = -\frac{Q/A}{\epsilon_0}\int  dx$

And do the integral:

$\Delta V = -\frac{Q/A}{\epsilon_0} x|_{A_x}^{B_x} = -\frac{Q/A}{\epsilon_0} * (B_x - A_x)$

Notice that the sign of this version of the answer is positive, just like the version above, because $(B_x - A_x)$ is negative.

```python
# We calculated E_x = -Q/A/e0 above
# for the points below I'm going to assume the origin of 
# my coordinate system is at location B
A_x = 1.7E-3
B_x = 0

voltage = -E*(B_x - A_x)

print (voltage)
```
