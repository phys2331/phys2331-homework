# The Problem

Here is a particular kind of mass spectrometer (see the figure). Carbon from the sample is ionized in the ion source at the left. The resulting singly ionized 12C+ and 14C+ ions have negligibly small initial velocities (and can be considered to be at rest). They are accelerated through the potential difference Δ𝑉1
. They then enter a region where the magnetic field has a fixed magnitude 𝐵=
 0.14 T. The ions pass through electric deflection plates that are 1 cm apart and have a potential difference Δ𝑉2
 that is adjusted so that the electric deflection and the magnetic deflection cancel each other for a particular isotope: one isotope goes straight through, and the other isotope is deflected and misses the entrance to the next section of the spectrometer. The distance from the entrance to the fixed ion detector is a distance of 𝑤=
 27 cm. There are controls that let you vary the accelerating potential Δ𝑉1
 and the deflection potential Δ𝑉2
 in order that only 12C+ or 14C+ ions go all the way through the system and reach the detector. You count each kind of ion for fixed times and thus determine the relative abundances. The various deflections insure that you count only the desired type of ion for a particular setting of the two voltages.

# The Problem Part 1 (a)

The accelerating plates are highlighted below.  This is the part of the accelerator that is responsible for giving the beam energy - for giving it a kick that gets the beam into the rest of the system.  Notice that this region only has an electric field - there is no magnetic field yet.



To understand which plate must be positive, we need to use $\vec{F} = q\vec{E}$.  The force needs to point to the right, towards the rest of the system.  The problem also tells us what the sign of $q$ is: <mark>12C+ and 14C+ ions</mark>.  This means the ions are positive.  

Since the force must point to the right and the particles are positively charged, the electric field must point to the right as well (since $\vec{F} = q\vec{E}$).  We can use PHeT to figure out how the charge must be arranged to get an electric field pointing to the right:

![acceleration_plate_configurations.png](images/acceleration_plate_configurations.png)

To get an electric field pointing to the right inside the acceleration plates, the positive charges need to be on the left plate.

# The Problem Part 1 (b)

The deflection plates are highlighted below.  This part of the accelerator selects a specfic particle velocity, so that all the particles entering the system have the same velocity.  Without this, it would be nearly impossible to analyze the data because we'd have many values of the speed.

Note that in the region of the deflection plates there is both an electric field (because of the plates) and a magnetic field (from the magnet, which is not shown in the diagram).

![Deflection Plates](images/mass_spectrometer_deflection.png)

The question asks, "Which deflection plate is positive, the top or bottom?" and this is really asking, "what must be the direction of the electric field to allow particles with speed $v$ to go through the deflection plates without bending?"

The only way for particles to move through this region without bending is if the net force on them is zero.  So we must have

$\vec{F}_{net} = \vec{F}_{magnetic} + \vec{F}_{electric} = q\vec{v}\times\vec{B} + q\vec{E} = 0$

Or

$\vec{v}\times\vec{B} = -\vec{E}$

So we can find $\vec{E}$ by knowing the directions of the velocity and the magnetic field.  The velocity is to the right, but I don't know the direction of the magnetic field - they ask for that in Part 1 (c)!  Let's do that problem and then come back to this one.

# The Problem Part 1 (c)

This part of the problem asks, "What is the direction of the magnetic field inside the spectrometer?" which I read as, "What direction does the magnetic field have to be to push a positive particle that's traveling to the right upwards?"

The force due to the magnetic field is

$\vec{F} = q\vec{v}\times\vec{B}$

My plan is to guess different values of $\vec{B}$ and see if they give me a force in the correct direction (upwards).  The problem helps me a little; the answer options are "into the page" and "out of the page".  So I only have to check two values.

```python
# I'm putting fake numbers in
# so I can figure out the direction of the force
# this value isn't accurate for q but it is the right sign!
q = +1

import numpy as np
v = np.array([1, 0, 0])
B_into_page = np.array([0, 0, -1])
B_outof_page = np.array([0, 0, 1])

Force1 = q*np.cross(v, B_into_page)
Force2 = q*np.cross(v, B_outof_page)
```

```python
print ("Force1 is ", Force1)
print ("Force2 is ", Force2)
```

# The Problem Part 1 (c) 

Force1 pushes the particle upwards, which is what we need to make the particle move the way the picture shows.

This force comes from using B pointing into the page, so B must point into the page.

# The Problem Part 1 (b)

Now that we know the direction of the magnetic field (into the page), we can finish part (b).

In our previous discussion of part (b) we said that the electric and magnetic fields had to sum to zero, which simplifies to

$\vec{v}\times\vec{B} = -\vec{E}$

or 

$\vec{E} = -\vec{v}\times\vec{B}$

As before, we need a coordinate system.  I'll let $+x$ point to the right, $+y$ point up, and $+z$ point out of the screen. 

Then I can take $-\vec{v}\times\vec{B}$ and find out the direction of the electric field.  Once I know the direction of the electric field, I can figure out where the positive charges must be for the deflection plates.

```python
import numpy as np

v = np.array([1,0,0])
B = np.array([0, 0, -1])

E_direction = -np.cross(v,B)

print (E_direction)
```

# The Problem Part 2

 Determine the appropriate numerical values of Δ𝑉1 and Δ𝑉2 for 12C. Carry out your intermediate calculations algebraically, so that you can use the algebraic results in the next part.
 
## $\Delta V1$

We can determine $\Delta V1$ because we know the radius of the particle's path in the magnetic field.  Since we know the magnetic field strength, the radius determines the velocity.  And that tells us $\Delta V1$ because this voltage is responsible for the speed through $Energy = q\Delta V1 = \frac{1}{2}mv^2$, or $\Delta V1 = \frac{1}{2q}mv^2$

So let's see how the radius determines the velocity.

$\vec{F} = m\vec{a} = q\vec{v}\times\vec{B}$

The particle is moving in a circle, so $\vec{a} = \frac{v^2}{r}\,pointing\,towards\,the\,center$.

Let's simplify the cross product $q\vec{v}\times\vec{B} = qvB\sin{\theta}\,pointing\,towards\,the\,center$.  In the diagram, the velocity of the particle and the magnetic field are at 90 degrees to each other.  So now $\vec{F} = m\vec{a} = q\vec{v}\times\vec{B}$ looks like

$m\frac{v^2}{r}\,pointing\,towards\,the\,center = qvB\,pointing\,towards\,the\,center$

I can take the magnitude of both sides of the equation and get a simple scalar equation:

$m\frac{v^2}{r} = qvB$

And now I can solve for $v$:

$v = \frac{qBr}{m}$

Now we can plug this velocity into the equation we got from equating the electric potential energy to the kinetic energy:

$\Delta V1 = \frac{1}{2q}mv^2 = \Delta V1 = \frac{1}{2q}m(\frac{qBr}{m})^2 = \frac{1}{2}\frac{qB^2r^2}{m}$

```python
# need the constants for mass, charge
from scipy import constants
# C12 has a mass of 12 AMU
# in other words,
# 1 mol of C12 has a mass of 12 g
# so 1 C12 has a very small mass:
m_12C_grams = 12/constants.Avogadro
m_12C_kg = m_12C_grams/1000
q = constants.elementary_charge
B = 0.14
r = (27E-2)/2

# There is a mistake in the formula below!
# can you fix it?
# The value should be about 1436 V
V1 = q*(B**2)*(r**2)/m_12C_kg

print (V1)
```

# The Problem Part 2 - $\Delta V2$

The voltage this part of the problem asks for is across the deflection plates, where particles of a certain velocity go straight through with no deflection:

![Deflection Plates](images/mass_spectrometer_deflection.png)

What this must mean is that the sum of the forces on the particles with a particular speed $v$ is zero, or

$\vec{F}_{net} = \vec{F}_{magnetic} + \vec{F}_{electric} = q\vec{v}\times\vec{B} + q\vec{E} = 0$

We know the magnetic field $B$ and as discussed in our calculation of $V1$, we know $v$ from the radius of the path, the strength of the magnetic field, and the charge of the ions.  This means that we can solve for the electric field $\vec{E}$ and from that we can calculate the voltage, provided we know the path.  The problem says, <mark>ions pass through electric deflection plates that are 1 cm apart</mark>, so we do have path information.

First, let's solve for the electric field.  I'm going to do this using vectors just so you can see how it's done.

$q\vec{v}\times\vec{B} + q\vec{E} = 0$

$q\langle v, 0, 0\rangle\times\langle 0, 0, -B\rangle + q\langle 0, -E, 0\rangle = 0$

$\langle 0, qvB, 0\rangle + q\langle 0, -E, 0\rangle = 0$

$\langle 0, qvB - qE, 0 \rangle = \langle 0, 0, 0\rangle$

Where in that last line I've replaced $0$ with $\langle 0, 0, 0\rangle$ since that's what we mean when we say a vector is zero.

The only component that's interesting is the y-component, so I'll write the equation for that single component:

$qvB - qE = 0$

Or $E = vB$.

From our work on $V1$, we got $v = \frac{qBr}{m}$, so we can substitute this into our expression for $E$.

$E = \frac{qB^2r}{m}$ and now we can calculate the electric field inside the deflection plates with quantities given in the problem.  But wait, we need the voltage across the deflection plates! For this we need to use

$\Delta V = -\int_A^B \vec{E}\cdot\vec{dl}$

In the image below I've shown my path and direction across the deflection plates.  I'll start from the bottom plate and walk to the top plate.  I could reverse this - that would change the sign of the voltage I calculate.  The problem is expecting me to take the absolute value of the voltage.

![Deflection Plates](images/mass_spectrometer_deflection_path.png)

$\Delta V2 = -\int_A^B \langle 0, -\frac{qB^2r}{m}, 0\rangle\cdot\langle dx, dy, dz\rangle$

$\Delta V2 = -\int_A^B 0dx -\frac{qB^2r}{m}dy + 0dz$

$\Delta V2 = \int_{y_{initial}}^{y{final}} \frac{qB^2r}{m}dy = \frac{qB^2r}{m} \int_{y_{initial}}^{y_{final}} dy = \frac{qB^2r}{m} y|_{y_{initial}}^{y_{final}} = \frac{qB^2r}{m}*(y_{final} - y_{initial})$



```python
# need the constants for mass, charge
from scipy import constants
# C12 has a mass of 12 AMU
# in other words,
# 1 mol of C12 has a mass of 12 g
# so 1 C12 has a very small mass:
m_12C_grams = 12/constants.Avogadro
m_12C_kg = m_12C_grams/1000
q = constants.elementary_charge
B = 0.14
r = (27E-2)/2
delta_y = 1E-2

# There is a mistake in the formula below!
# can you fix it?
V2 = (q*(B**2)*r/m_12C_kg) * delta_y

print (V2)
```
