# The Problem
At one time, concern was raised about the possible health effects of the small alternating (60 hertz) magnetic fields created by electric currents, in houses and near power lines. In a house, most wires carry a maximum of 15 amperes (there are 15 ampere fuses that melt and break the circuit if this current is exceeded). The two wires in a home power cord are about =
 3 millimeters apart, as shown in the figure, and at any instant they carry currents in opposite directions (both of which change direction 60 times per second).
 
Calculate the maximum magnitude of the alternating magnetic field, =
 66 cm away from the center of a long straight power cord that carries a current of 15 amperes. Both wires are at the same height as the observation location.
 



# The Solution

The appropriate equation to use here is the magnetic field due to a long wire,

$B = \frac{\mu_0 I}{2\pi r}$

This is the equation you get when you add up all the contributions from the Biot-Savart law, $\vec{\Delta B} = \frac{\mu_0}{4\pi}\frac{I\vec{\Delta l}\times\hat{r}}{r^2}$, from a straight wire.

Notice that the equation $B = \frac{\mu_0 I}{2\pi r}$ only gives you magnitude information.  You are expected to use pictures like the one below to figure out what the direction is.  If you're familiar with polar coordinates, the direction can be described as $\hat{\theta}$.  Since the current through each wire points opposite, the magnetic fields due to each wire must be pointing opposite each other.



```python
from scipy import constants
```

```python
mu0, mu0_units, mu0_uncertainty = constants.physical_constants['vacuum mag. permeability']
```

```python
I_max = 15
```

```python
d1 = 66E-2
B1 = mu0/(2*constants.pi)* I_max/d1

print (B1)
```

```python
s = 3E-3
d2 = d1-s

B2 = -mu0/(2*constants.pi)* I_max/d2

print (B2)
```

```python
print (B1+B2)
```

```python
import numpy as np

print (np.abs(B1+B2))
```
